## Import necessary libraries

```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

## Import data 

```python
concrete = pd.read_csv("concrete_data.csv")
concrete.head(5)
```

```python
columns = concrete.columns
```

## Normalize the dataset by sutracting mean and dividing by standard deviation

```python
import numpy as np
def normalize_df(df,columns):
    for i in columns:
        df[i] = (df[i] - np.mean(df[i]))/ np.sqrt(np.var(df[i]))
    return df
```

```python
Ndf = normalize_df(concrete,columns)
```

## Split the data into train and test

```python
X = Ndf.loc[: ,Ndf.columns != 'Strength']
y = Ndf['Strength']
X_train, X_test, y_train, y_test = train_test_split(X ,y , test_size = 0.30, random_state = 62)
```

```python
y_test.shape
```

```python
n_cols = X.shape[1]
```

## Train the model on Training data using 100 epochs

```python
model = Sequential()
model.add(Dense(10, activation = 'relu', input_shape = (n_cols,)))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense( 1, activation = 'relu'))
model.compile(optimizer ='adam', loss = 'mean_squared_error')
model.fit(X_train, y_train, epochs = 100)
```

## Predict on the test data

```python
predict = model.predict(X_test)
```

```python
predict.shape
```

## Calculate the mean squared error

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, predict)
print(mse)
```

## Calculating mse 50 times 

```python
mse = []
for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(X ,y , test_size = 0.30)
    model = Sequential()
    model.add(Dense(10, activation = 'relu', input_shape = (n_cols,)))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense( 1, activation = 'relu'))
    model.compile(optimizer ='adam', loss = 'mean_squared_error')
    model.fit(X_train, y_train,epochs = 100)
    predict = model.predict(X_test)
    mse.append(mean_squared_error(y_test, predict))
```

## Calculate the mean and standard deviation of mean squared error

```python
## Mean of mse after building the model on normalized data
mean_mse = np.mean(mse)
mean_mse
```

```python
## SD of mse after building the model on normalized data
variance_mse = np.var(mse)
sd_mse = np.sqrt(variance_mse)
sd_mse
```

## If the model is built on Normalized dataset and the number of epochs is increased as well as number of dense layer are increased, the mean mse is lower meaning the model is able to learn even better but in this case adding more layers in not improving the mse that much.
